```python
from src.utils.plotter import Plotter

plotter = Plotter(
    title= "", 
    y_axis_start_at_zero=False,
    columns=1
)
print("Log Length: ", LOG["0"].sum())

to_plot = list(LOG["0"].rolling(10).mean())
plotter.add_subplot(
    [
        (f"Window Size", to_plot),
    ],
    y_label="No. of events",
    x_range=list(
        range(len(to_plot))
    ),
)

fig = plotter.plot(x_label="Window Index", fig_height_per_plot=5, disable_show=True)
```
